```
import pandas as pd
import numpy as np
import plotly.io as pio
import plotly.express as px
import plotly.figure_factory as ff
import plotly.express as px
import io
import requests
```

```
aqrv = 'https://opendata.ecdc.europa.eu/covid19/casedistribution/csv'
df = pd.read_csv(aqrv, encoding="utf-8") #parse_dates=['dateRep'], infer_datetime_format=True,
```

```
df.head(10)
```

```
df.info()
```

```
df['dateRep'] = pd.to_datetime(df['dateRep'], format="%d/%m/%Y")
```

```
grafi = df.groupby('dateRep')['cases', 'deaths'].sum().reset_index()
```

```
grafi = grafi.melt(id_vars="dateRep", value_vars=['cases', 'deaths'],
                 var_name='Case', value_name='Count')
```

```
grafi.info()
```

```
from plotly.offline import plot, iplot, init_notebook_mode
init_notebook_mode(connected=True)

pio.renderers
pio.renderers.default = "colab"
```

```
cases = '#21bf73'
deaths = '#ff2e63'
```

```
import plotly.express as px
fig = px.area(grafi, 
              x="dateRep", 
              y="Count", 
              color='Case', 
              height=600,
              title='Casos no mundo ao longo do tempo',
              color_discrete_sequence = [cases, deaths])
fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```

```
df_agrupado = df.groupby(['dateRep', 'countriesAndTerritories'])['cases', 'deaths'].sum().reset_index()
```

```
df_agrupado['countriesAndTerritories'] = df_agrupado['countriesAndTerritories'].replace(['United_States_of_America'], 'US')
```

```
fig = px.choropleth(df_agrupado,                                          # casos agrupados por país
                    locations="countriesAndTerritories",                             # definindo as regiões no mapa
                    locationmode='country names',# define o modo de localização para todas regiões
                    color=np.log(df_agrupado["cases"]),# define a cor pelo o valor de casos confirmados (aplica o log)
                    hover_name='countriesAndTerritories',   # define o texto interativo com o nome da região
                    hover_data=["cases", "deaths"], # define o texto interativo com o numero de casos confirmasos e mortes
                    animation_frame=df_agrupado["dateRep"].dt.strftime('%d-%m-%Y'),   # define o animate_frame com as datas
                    title='Casos ao longo do tempo',         # define título
                    color_continuous_scale=px.colors.sequential.Magenta)    # define a paleta de cores
fig.update_layout(autosize=False, width=1200, height=800)  # define tamanho da figura
fig.show()
```
