```
import matplotlib.pyplot as plt
import pandas as pd
```

```
df = pd.read_csv("/content/Pilotos.csv", sep=",")
df.head()
```

```
df.describe()
```

```
df_podios=df.loc[(df['Pódios'] > 0)]
df_podios.head()
```

```
plt.style.use("classic")

piloto = df_podios.sort_values(by=["Pódios"], ascending=False)["Piloto"]
podios = df_podios.sort_values(by=["Pódios"], ascending=False)["Pódios"]
plt.bar(piloto,podios)
plt.title("Pilotos que Obtiveram Pódios na Temporada")
plt.xticks(rotation=90)
plt.xlabel('Pilotos')
plt.ylabel('Pódios')
plt.show()
```
